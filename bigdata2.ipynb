```python
df = pd.merge(prices_df, items_df, on=["item_id"],how="inner")
print(df.sample(5))
```

```python
region_prices = df.copy()
average_prices = region_prices.groupby(['region']).agg({'price': ['median','mean']})
average_prices = average_prices.reset_index()
average_prices.columns = ['Region','median','mean']
average_prices
```

```python
df = pd.merge(average_prices, regions_df, left_on ='Region', right_on='region_n',how='inner')
df
```

```python
cols = ['region', 'median','mean']
cleandf= df[cols]
```

```python
melted_df = cleandf.melt(id_vars='region', var_name='metric',value_name = 'price')

melted_df.to_csv('bigdata2.csv')
```
