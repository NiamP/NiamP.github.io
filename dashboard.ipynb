```python
import requests 
import pandas as pd
```

```python
URLs = {
    "Employment rate aged 16-24" : "/employmentandemployeetypes/timeseries/lf24/lms",
    'Female employment rate' : "/employmentandemployeetypes/timeseries/lf25/lms",
    'Male employment rate':"/employmentandemployeetypes/timeseries/mgsv/lms",
    'Average actual weekly hours worked for full time employees':'/earningsandworkinghours/timeseries/ybuy/lms',
    'Seasonally adjusted regular pay':'/earningsandworkinghours/timeseries/kai7/emp',
    'Temporary Employees could not find permanent job':'/employmentandemployeetypes/timeseries/ycci/lms',
    'UK Job Vacancies': '/employmentandemployeetypes/timeseries/jp9z/unem',
    'Output per hour worked percentage change': '/labourproductivity/timeseries/lzvd/prdy',
    'Public Sector Workforce size': '/publicsectorpersonnel/timeseries/g7g3/pse',
    'Time lost due to strike action': '/employmentandemployeetypes/timeseries/bbfw/lms'
}   


alldata=[]
for dataset_name, URL in URLs.items():
    request = requests.get(f'https://api.beta.ons.gov.uk/v1/data?uri=/employmentandlabourmarket/peopleinwork{URL}')
    jsondata = request.json()

    if 'years' in jsondata and jsondata['years']:
        for item in jsondata['years']:
            alldata.append({
                "dataset": dataset_name,
                "date": item['date'],
                "value": item['value'],
                })
    
    elif 'quarters' in jsondata and jsondata['quarters']:
        for item in jsondata['quarters']:
            alldata.append({
                "dataset": dataset_name,
                "date": item["date"],
                "value": item["value"],
            })

print(alldata)

```

```python
data1 = requests.get('https://api.beta.ons.gov.uk/v1/data?uri=/employmentandlabourmarket/peopleinwork/employmentandemployeetypes/timeseries/bbfw/lms')
print(data1.json())
```

```python
df = pd.DataFrame(alldata)

print(df.head())

```
