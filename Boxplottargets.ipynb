```python
import pandas as pd
```

```python
data = pd.read_excel('/Users/niampopat/Library/CloudStorage/OneDrive-UniversityofBristol/Documents/Data science for Economists/SBTiMonitoringReport2022TargetProgressData.xlsx',sheet_name = 'Target Progress')

data = data[['Company name', 'Region ', 'Target value', 'Target year','Target','Sector']]

data
```

```python
data['Target value'] = pd.to_numeric(data['Target value'], errors='coerce')


data = data[data['Target'].str.contains('Near-term', case=False, na=False)]
data = data.reset_index(drop=True)
data
```

```python
data = data.groupby('Company name').apply(lambda x: x.nlargest(1, 'Target value')).reset_index(drop=True)
data
```

```python
cleandf = data[['Company name','Target value','Sector']]
cleandf = cleandf[cleandf['Target value']<=1]
cleandf
```

```python
cleandf.loc[cleandf.index, 'Target value'] = cleandf['Target value'].apply(lambda x: x * 100)
sectorcount = cleandf['Sector'].value_counts()
sectorstokeep = sectorcount[sectorcount>=10].index
cleandf = cleandf[cleandf['Sector'].isin(sectorstokeep)]
sectorcount2 = cleandf['Sector'].value_counts()
cleandf

```

```python
cleandf['Target value'] = cleandf['Target value'].round(2)
print(cleandf)
```
